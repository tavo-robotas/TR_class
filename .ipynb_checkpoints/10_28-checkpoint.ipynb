## Kintamieji

Python programavimo kalboje norint sukurti kintamaj nereikia nurodyti atskirai kokio tipo duomenys saugos is naujai sukurtas kintamasis. Tai reikia, pvz:

```python
mano_kitamasis = 5
```

Python programavimo kalboje yra daug paruot funkcij kurios padeda atlikti visokius udavinius, tame tarpe ir nustatyti kintamojo typ, tam tisklui mes naudosime funkcij `type()`

```python
type(mano_kitamasis)
```


Kaip matote reikia tiktai duoti kintamajam vard ir priskirti vert, python automatikai nustato kas tai per duomen tipoas. Yra tiktai keletas taisykli kaip reikia vadinti kintamuosius. Pirmiausia turite parinkti tinkam jiems vard ! Idealiu atveju kintamj pavadinimai yra trumpi ir aikiai juos apibdinantys. Jie taip pat turi atitikti kelet reikalavim:
 - Juose negali bti tarp (pvz., pirmas kintamasis yra neleistina forma)
 - Juose gali bti tik raidi, negalima naudoti skaii ir ar t keist simboli: <code>**!@#$%^& ir t.t**<code> 
 - Jie turi prasidti tiktai raide arba pabraukimu (pvz., 1_var neleistina forma):
 - Kintamasis turi bti i vieno odio, taiau jeigu reikia dvej tada juos sujungiame su _ (underscore) arba taikome taip vadinama kupranugario format, pvz:
   - mano_kitamojo_pavadinimas_yra_labai_ilgas
   - manoKintamjojoPavadinimasYraLabaiIlgas

Sukurdami kintamj, turite naudoti lygybs enkl <code>**=**</code>, tam kad priskirtumte jam norim reikm.
Visada galite pavelgti  kintamojo priskirt reikm naudodami print() ir skliausteliuose raydami kintamojo pavadinim.

```python
kitas_kintamasis = 'cia bus mano tekstas'
```

```python
print(kitas_kintamasis)
```

```python
type(kitas_kintamasis)
```

```python
dar_vienas = True
```

```python
type(dar_vienas)
```

```python
kintamasis = 5.5
```

```python
type(kintamasis)
```

Kintamuosius galima perayti naudojant t pati vard:

```python
# kintamojo mano_kin reikm  1
mano_kin = 1

# Ispausdinkite mano_kin priskirt reikm
print("pradin reikm:", mano_kin)

# Pakeiskite kintamojo reikm  100
mano_kin = 100

# Ispausdinkite nauj reikm, priskirt mano_kin
print("pakeista  reikm:",mano_kin)
```

Atkreipkite dmes, kad paprastai, kai apibriate kintamj kodo langelyje, visi tolesni kodo langeliai taip pat turi prieig prie io kintamojo. Pavyzdiui, mes naudojame kit kodo langel, kad pasiektume my_kim (i aukiau esanio kodo langelio) ir suma (i dar ankstesnio kodo langelio) reikmes. <code>**Kintamasis yra kompiuterio atmintyje !**</code>

```python
mano_kin
```

Galima naudojanti kelis kintamuosius ikarto tai yra prasta praktika.
Kitame kodo langelyje apskaiiuojame sekundi skaii per ketverius metus. iam skaiiavimui naudojami penki vesties duomenys.

```python
# Kintamieji
metai = 4
dienos = 365 
valandos = 24
minutes = 60
sekundes = 60

# Suskaiiuokime kiek sekundi yra keturiuose metuose
viso = sekundes * minutes * valandos * dienos * metai
print(viso)
```

Kaip apskaiiuota aukiau, per ketverius metus praeina <code>**126144000 sekundi**</code>
