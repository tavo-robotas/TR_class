## Kintamieji

Python programavimo kalboje norint sukurti kintamajį nereikia nurodyti atskirai kokio tipo duomenys saugos šis naujai sukurtas kintamasis. Tai reiškia, pvz:

```python
mano_kitamasis = 5
```

Python programavimo kalboje yra daug paruoštų funkcijų kurios padeda atlikti visokius uždavinius, tame tarpe ir nustatyti kintamojo typą, tam tisklui mes naudosime funkciją `type()`

```python
type(mano_kitamasis)
```


Kaip matote reikia tiktai duoti kintamajam vardą ir priskirti vertę, python automatiškai nustato kas tai per duomenų tipoas. Yra tiktai keletas taisyklių kaip reikia vadinti kintamuosius. Pirmiausia turite parinkti tinkamą jiems vardą ! Idealiu atveju kintamųjų pavadinimai yra trumpi ir aiškiai juos apibūdinantys. Jie taip pat turi atitikti keletą reikalavimų:
 - Juose negali būti tarpų (pvz., pirmas kintamasis yra neleistina forma)
 - Juose gali būti tik raidžių, negalima naudoti skaičių ir ar tų keistų simbolių: <code>**!@#$%^& ir t.t**<code> 
 - Jie turi prasidėti tiktai raide arba pabraukimu (pvz., 1_var neleistina forma):
 - Kintamasis turi būti iš vieno žodžio, tačiau jeigu reikia dvejų tada juos sujungiame su _ (underscore) arba taikome taip vadinama kupranugario formatą, pvz:
   - mano_kitamojo_pavadinimas_yra_labai_ilgas
   - manoKintamjojoPavadinimasYraLabaiIlgas

Sukurdami kintamąjį, turite naudoti lygybės ženklą <code>**=**</code>, tam kad priskirtumėte jam norimą reikšmę.
Visada galite pažvelgti į kintamojo priskirtą reikšmę naudodami print() ir skliausteliuose įrašydami kintamojo pavadinimą.

```python
kitas_kintamasis = 'cia bus mano tekstas'
```

```python
print(kitas_kintamasis)
```

```python
type(kitas_kintamasis)
```

```python
dar_vienas = True
```

```python
type(dar_vienas)
```

```python
kintamasis = 5.5
```

```python
type(kintamasis)
```

Kintamuosius galima perašyti naudojant tą pati vardą:

```python
# kintamojo mano_kin reikšmę  1
mano_kin = 1

# Išspausdinkite mano_kin priskirtą reikšmę
print("pradinė reikšmė:", mano_kin)

# Pakeiskite kintamojo reikšmę į 100
mano_kin = 100

# Išspausdinkite naują reikšmę, priskirtą mano_kin
print("pakeista  reikšmė:",mano_kin)
```

Atkreipkite dėmesį, kad paprastai, kai apibrėžiate kintamąjį kodo langelyje, visi tolesni kodo langeliai taip pat turi prieigą prie šio kintamojo. Pavyzdžiui, mes naudojame kitą kodo langelį, kad pasiektume my_kim (iš aukščiau esančio kodo langelio) ir suma (iš dar ankstesnio kodo langelio) reikšmes. <code>**Kintamasis yra kompiuterio atmintyje !**</code>

```python
mano_kin
```

Galima naudojanti kelis kintamuosius iškarto tai yra įprasta praktika.
Kitame kodo langelyje apskaičiuojame sekundžių skaičių per ketverius metus. Šiam skaičiavimui naudojami penki įvesties duomenys.

```python
# Kintamieji
metai = 4
dienos = 365 
valandos = 24
minutes = 60
sekundes = 60

# Suskaičiuokime kiek sekundžių yra keturiuose metuose
viso = sekundes * minutes * valandos * dienos * metai
print(viso)
```

Kaip apskaičiuota aukščiau, per ketverius metus praeina <code>**126144000 sekundžių**</code>
